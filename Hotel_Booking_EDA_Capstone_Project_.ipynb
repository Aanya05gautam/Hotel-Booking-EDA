1 . ***`Cancellation Trends:`***

10 . ***`Seasonality:`***

11 . ***`Business Implications and Recommendations:`***
Cancellation Reduction: By identifying factors influencing cancellations—like lead time, guest type, and deposit policy—the hotel can customize booking terms, introduce non-refundable deposits selectively, and offer early confirmation incentives.

## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Check null values
df.isnull().sum()
```

```python
# Fill missing 'children' with 0 assuming missing means none
df['children'] = df['children'].fillna(0)
```

```python
# Fill 'agent' with 0 (assuming missing agent means direct booking or unknown)
df['agent'] = df['agent'].fillna(0)
```

```python
# Fill 'country' with mode
df['country'] = df['country'].fillna(df['country'].mode()[0])
```

```python
# Add a few useful columns:
# 1. Total stay in nights
df['total_stay'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
# Total guests
df['total_guests'] = df['adults'] + df['children'] + df['babies']

# Create a mapping for month names to numbers
month_map = {'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
             'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12}

# Create arrival_month_num column
df['arrival_month_num'] = df['arrival_date_month'].map(month_map)

# Create full arrival date
df['arrival_date'] = pd.to_datetime(df['arrival_date_year'].astype(str) + '-' + df['arrival_month_num'].astype(str) + '-' + df['arrival_date_day_of_month'].astype(str))
```

#### Chart - 1   Bar Plot – Hotel Type vs Cancellation Rate

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate cancellation rate per hotel (in percentage)
cancel_rate = df.groupby('hotel')['is_canceled'].mean().reset_index()
cancel_rate['cancellation_rate_percent'] = cancel_rate['is_canceled'] * 100

plt.figure(figsize=(8,6))
# Use 'hue' with legend=False to avoid the FutureWarning
bars = sns.barplot(x='hotel', y='cancellation_rate_percent', data=cancel_rate, hue='hotel', palette='Set2', legend=False)

plt.title('Cancellation Rate by Hotel Type (%)', fontsize=16)
plt.ylabel('Cancellation Rate (%)')
plt.xlabel('Hotel Type')

# Add percentage labels on top of bars
for bar in bars.patches:
    height = bar.get_height()
    bars.text(bar.get_x() + bar.get_width()/2, height + 1, f'{height:.1f}%',
              ha='center', va='bottom', fontsize=12)

plt.ylim(0, cancel_rate['cancellation_rate_percent'].max() + 10)
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

Answer:

Bar plots are perfect for comparing categorical variables like hotel type.
